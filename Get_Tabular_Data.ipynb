### 6.4.1. Getting Data with the CSV Library

The comma separated values (CSV) format is a common import and export format for spreadsheets, databases, and programming languages. CSVs are for the transfer of data between systems and are readable by many applications and programs, including Excel. The CSV Library provides functions to both read and write CSV files. Below are a couple of terms that may be unfamiliar but are helpful to understand.

* `delimiter` is a character or string that separates individual pieces of data.
* `quotechar` wraps data that contains your delimiter character, allowing it to be read as one piece of data.

Whenever we work with libraries in Python, we need to import them into our script. We only have to do this once inside the current kernel, but if we restart the kernel we need to run the import statement again.

````python
import csv
````

Now that we've imported the library, we can use `help(csv)` to explore all of its features.

````python
help(csv)
````

#### 6.4.1.1. Reading a Basic CSV File with `csv.reader()`

`csv.reader()` returns an object that all the data from each row in a CSV file. To retain the data, it must be appended to a list for you to manipulate later in the code. These objects are good for storing data but are not easy to work with. Because of this limitation, programmers typically push the data from the `.reader()` object into a standard Python list.

The code below, reads in a new CSV file stored on your computer. The variable `animal` stores the object created by `csv.reader()`. We then use a `for` loop to step through each row of data and print it to the screen.

````python
with open('data/animal/animal_info_land.csv') as csvfile:
    
    animal = csv.reader(csvfile)
    
    for row in animal:
        print(row)
````

Can you see the tabular structure? If you're curious what the `.reader()` object looks like, run the code below.

````python
animal
````

Remember that printing data is not the same as storing it in memory. In order to save the information for further use, we need to append each row to a list.

````python
land_animals = []

with open('data/animal/animal_info_land.csv') as csvfile:
    
    animal = csv.reader(csvfile)
    
    for row in animal:
        land_animals.append(row)

land_animals
````

#### 6.4.1.2. Writing to a CSV File
Writer objects can take data and write it to a file in CSV format. We create them like this:

```python
csv.writer(csvfile)
```

Before we write our data out, let's change a piece of it. We do this for two reasons: 1) this will show that we've actually written the data we have, not just copied an old file, and 2) our new data value will have our delimiter character (`','`) in it, so this will demonstrate the use of quotechars. Here, we'll edit the third item in the last column of our table.

````python
land_animals[-1][2] = 'Grass, Bananas'
land_animals
````

Writing with the CSV Library is very similar to using the standard `.write()` method. The syntax is slightly different, because we first need to create a `.writer()` object before we can write any data. The `.writer()` object takes the file stream as an argument. To actually write data, we call the object and its method, `.writerows()`, and pass in a list of lists.

NOTE: The `newline=''` argument is important for writing CSVs that we intend to open in Excel. Try taking it out and see how the output changes.

````python
with open('data/animal/animal_output.csv', mode='w', newline='') as csvfile:
    
    writer = csv.writer(csvfile)
    
    writer.writerows(land_animals) 
````

#### 6.4.1.3. Using `.DictReader()` to read a CSV File

Another way to interpret CSV data is as a list of dictionaries. Let's look at an example. Here is some sample CSV data:

```python
csv_data = '''Name,Weight,Diet,Average Lifespan\n
              Giraffe,2600,Shrubbery,25\n
              Moose,1300,Shrubbery,20\n
              Giant Panda,220,Bamboo,20\n'''
```   

This is the same as the following table:
                  
| Name       | Weight     | Diet      | Average Lifespan    |
| :--------- | :---------: | :------:   | :-------------------:|
| Giraffe    | 2600       | Shrubbery | 25	                |
| Moose      | 1300       | Shrubbery | 20	                |
| Giant Panda| 220        | Bamboo    | 20               	|

This same information is presented as a list of dictionaries below:

```python
csv_data = [{'Average Lifespan': '25',
             'Diet': 'Shrubbery',
             'Name': 'Giraffe',
             'Weight': '2600'},
            {'Average Lifespan': '20',
             'Diet': 'Shrubbery',
             'Name': 'Moose',
             'Weight': '1300'},
            {'Average Lifespan': '20',
             'Diet': 'Bamboo',
             'Name': 'Giant Panda',
             'Weight': '220'}]
```

Can you see the different ways each format captures the same information?

`csv.DictReader()` is similar to `csv.reader()` in function but returns a slightly different object. This object formats the data in each row as a dictionary, using the first row of data as the keys. To retain the data outside of `.DictReader()` object, append each dictionary to a list.

````python
land_animals_dict = []

with open('data/animal/animal_info_land.csv') as csvfile:
    
    animal = csv.DictReader(csvfile)
    
    for row in animal:
        land_animals_dict.append(row)

land_animals_dict
````

One reason to use dictionaries over lists is that the data is more intuitively stored. Rather than having to know what order your data values are stored in, as with a list, you can call individual pieces of each row with their column name, which usually describes what the data is about. The dictionaries above look a little different from standard Python dictionaries, but they work basically the same. Below, we extract the `Name` value from the second row in our table.

````python
land_animals_dict[1]['Name']
````

<hr>
#### 6.4.1.4. Using `.DictWriter()` to write to a CSV File

Writer objects can take a data collection, convert it into a string, and write it to a given file. `csv.DictWriter()` is similar to `csv.writer()`, but takes a list of dictionaries as input rather than a list of lists. When we call `.DictWriter()`, we also need to pass the fieldnames, or column names, for our data.

Again, let's alter our data slightly to demonstrate that this is really working. Below, we add a row to our dataset.

````python
tiger_data = {'Name': 'Tiger', 'Diet': 'Deer', 'Average Lifespan': 10, 'Weight': 450}

land_animals_dict.append(tiger_data)
land_animals_dict
````

Now we will write the new dataset to a file. First, we specify our field names as a list of strings. Then we pass that list along with our file stream (created, as usual, with the `with open()` statement) to `.DictWriter()`. Once we have the writer object, we use it to write the table header (`.writeheader()`) and write the rows of the table (`.writerows()`).

````python
with open('data/animal/animal_dict_output.csv', mode='w', newline='') as csvfile:
    
    fieldnames = ['Name', 'Weight', 'Diet', 'Average Lifespan']
    
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    
    writer.writerows(land_animals_dict)
````

Find this file in your file system and open it. Does it look how you expected it to? Let's read it back into our script to make sure it was written successfully.

````python
animals_list_of_dicts = []

with open('data/animal/animal_dict_output.csv') as csvfile:
    
    animal = csv.DictReader(csvfile)
    
    for row in animal:
        animals_list_of_dicts.append(row)

animals_list_of_dicts
````

### 6.4.2. The Glob Library: How to Read A LOT of Data into a Python Script
<hr>
#### References:
* [Python 3.4: glob](https://docs.python.org/3.4/library/glob.html)
<hr>

You may have noticed that with each lesson we are increasing the amount of data at your fingertips. In Lesson 5 we processed a number of resume files using a list of all the file names. We were able to manually type out this list because it consisted of only ten files, but what if we want to process a thousand files? 

When typing every file name manually is out of the question, use the Glob Library. The Glob Library allows you to grab all of the file names that match a specified pattern. Yes, Glob allows you to load all of the data files in your Documents folder for example, but doing so might cause your computer to crash. It's smart to be cognizant of the amount of data you are working with and your computational resources.

The Glob Module finds all the pathnames matching a specified pattern. The syntax for implementing glob is below.

```python
files = glob.glob('folder1/folder2/*.*')
```

As always, our use of glob requires us to first import the library.

````python
import glob
````

Now, we'll call the Glob Library's `.glob()` method. The only argument it takes is a string that supplies the pattern to match file names to. To allow `.glob()` to match files with different names, we use a wildcard character (`'*'`). The asterisk will match any amount of characters of any kind.

````python
files = glob.glob('data/animal/*.*')
files
````

Take note of the output. We passed the `.glob()` method the pattern `'data/animal/*.*'`, which matches any file in the `data/animal/` folder with a period (`'.'`) in it. In return, we get a list of strings, each string corresponding to a file on our computer.

As we mentioned, the `'*'` is a catch-all for any string of characters. As such, the pattern `'*.*'` matches any file with a period in it. All the file names below match this pattern.

```python        
'animal.csv'
'data.json'
'info.xlsx'
```

A more specific wildcard expression would be `'*.csv'` which would only collect files with a `.csv` extension. Since we're focusing on CSV files in this lesson, let's change our previous pattern to match only CSV files.

````python
files = glob.glob('data/animal/*.csv')
files
````

NOTE: The double backslash (`'\\'`) is the way `.glob()` separates folders and files when it runs in a Windows environment. In Python, `'\\'` and `'/'` are interchangeable as path separators.

#### 6.4.2.2. Hard-Coding vs. Programmatic Coding

Now that we're using strings to connect our Python scripts to files on our computer, let's discuss the concept of _hard-coding_. Hard-coding is when you type explicit values in your code. So far, all our paths and patterns have been hard-coded. *Programmatic coding*, on the other hand, is using variables wherever possible in your functions and method calls in lieu of explicit values.

When writing scripts for real world use cases, you need to think about how someone else could run your code on their machine.  Ask yourself: How easily can my code be applied in a slightly different context? By writing programmatic code, you make your scripts useful to others as well as yourself.

For example, let's say we write a script that approximates the diameter and area of a circle given the circle's radius. We might pick a specific radius to start out with to make sure our code works the way we want it to. Given a circle of radius 150, our code could look something like this:

````python
diameter = 150 * 3.14159 * 2

area = 150**2 * 3.14159

print('diameter:', diameter)
print('area:', area)
````

One thing we could do to make this code more programmatic is to store our approximation of pi (`3.14159`) in a variable, and use that variable every time we use pi in our code. Doing so would give us the following:

````python
PI = 3.14159

diameter = 150 * PI * 2

area = 150**2 * PI

print('diameter:', diameter)
print('area:', area)
````

Now this might not seem like a big improvement, but let's say in the future we want to make our approximation of pi more accurate. Instead of having to go in and change two values in the code (one in each of our equations) we only need to change it once, in the variable assignment, and the new value is used everywhere we call that variable. 

````python
PI = 3.14159265359

diameter = 150 * PI * 2

area = 150**2 * PI

print('diameter:', diameter)
print('area:', area)
````

Imagine this is part of a larger, more complicated script in which you use pi two dozen times. Coding programmatically by storing the value to a variable makes us much faster and much less error-prone in the long run.

Now our code is a little more robust thanks to our smart use of a single variable. But what if we wanted to find the diameter and area of a circle with a radius of 7500? If we're hard-coding, we'd have to go in a swap in `7500` everywhere we see the value `150` in our code above.

````python
PI = 3.14159265359

diameter = 7500 * PI * 2

area = 7500**2 * PI

print('diameter:', diameter)
print('area:', area)
````

This is doable with only two lines of code to change, but what if our script used that radius value in a hundred other places? Do you trust yourself to remember to swap every single value in your code? Better question: Why even bother if you can use a variable to do it for you?

````python
PI = 3.14159265359
radius = 7500

diameter = radius * PI * 2

area = radius**2 * PI

print('diameter:', diameter)
print('area:', area)
````

Now we can calculate the area and diameter of *any* circle just by changing a single value. Go back to the beginning of this section and notice how our code has changed. We've managed to make it more readable, more understandable, and more generalizable by using a couple of variables in the place of hard-coded values. We've also made editing and debugging easier. Coding programmatically is a good habit to build as you write your own code.

In the code cells below, we will work out the process for reading in and cleaning a FlightRadar 24 file.

Step 1. Read a file in using the `csv.DictReader()` method:

````python
track = []

with open('data/fr24/RA64520_ccbc1f8.csv') as csvfile:
    
    d_reader = csv.DictReader(csvfile)
    
    for row in d_reader:
        track.append(row)
        
track
````

Step 2. `.DictReader()` gave us a list of dictionaries. Take a look at the first dictionary in the list and examine the data more closely.

````python
track[0]
````

It looks like the `Position` column contains strings with latitude and longitude information. This data is very useful, but hard to work with in its current form. We'd like to isolate the latitudes and longitudes and turn them into numbers.

Step 3. Split the `Position` column into two separate columns titled `'Latitude'` and `'Longitude'`, and cast the data to floats. Do this by splitting the string in the `Position` column on the comma character.

````python
track[0]['Position'].split(',')
````

Now let's iterate over all the rows in our data. For each row, we execute the commands to create the `Latitude` and `Longitude` columns and delete the `Position` column. Then we inspect the first row again to make sure it looks good.

````python
for row in track:
    row['Latitude'] = float(row['Position'].split(',')[0])
    
    row['Longitude'] = float(row['Position'].split(',')[1])
    
    del row['Position']
    
track[0]
````

Step 4. Looks like we still have numeric data stored as strings. Format the rows `'Altitude'`, `'Speed'`, and `'Direction'` as integers.

````python
for row in track:
    row['Altitude'] = int(row['Altitude'])
    
    row['Speed'] = int(row['Speed'])
    
    row['Direction'] = int(row['Direction'])
    
track[0]
````

The difference is subtle, but you can see that the quotation marks went away in the three columns we edited. This tells us we now have numeric values, not strings.

Step 5. Add a row containing the name of the file where the data came from (this will be helpful later when we look at many files with a similar structure). Let's pull the value out of the filename above, and save it into a new entry in each row. 

````python
for row in track:
    row['ID'] = 'RA64520_ccbc1f8'
    
track[0]
````

Now that we have all the steps to clean our data, let's put them all in one cell. All we need to do to read and clean a different file is change the file name we pass into the `open()` statement.

````python
track = []

with open('data/fr24/RA64520_ccbc1f8.csv') as csvfile:
    d_reader = csv.DictReader(csvfile)
    for row in d_reader:
        track.append(row)

for row in track:
    row['Latitude'] = float(row['Position'].split(',')[0])
    row['Longitude'] = float(row['Position'].split(',')[1])
    del row['Position']
    row['Altitude'] = int(row['Altitude'])
    row['Speed'] = int(row['Speed'])
    row['Direction'] = int(row['Direction'])
    row['ID'] = 'RA64520_ccbc1f8'
````

````python
## INSTRUCTION SOLUTION(S) ##

def read_csv(filename):
    """Reads in and cleans FR_24 data.
    
    Args:
        filename -- (str) a file name including the relative path to the file
        
    Returns:
        track -- (list of dict) the cleaned data
    """
    
    track = []
    
    with open(filename) as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            track.append(row)
            
    for row in track:
        row['Latitude'] = float(row['Position'].split(',')[0])
        row['Longitude'] = float(row['Position'].split(',')[1])
        del row['Position']
        row['Altitude'] = int(row['Altitude'])
        row['Speed'] = int(row['Speed'])
        row['Direction'] = int(row['Direction'])
        row['ID'] = filename[10:-4]
        
    return track


read_csv('data/fr24/RA64520_ccbc1f8.csv')
````

````python
## YOUR CODE GOES HERE ##
````

````python
## INSTRUCTION SOLUTION(S) ##
callsigns = {}
    
for track in flighttracks:
    
    current_callsign = track[0]['Callsign']
    
    if current_callsign in callsigns:
        callsigns[current_callsign] += len(track)
        
    else:
        callsigns[current_callsign] = len(track)

callsigns
````

Count all of the area codes found in the `'Phone Number'` column (2nd Column).

Phone Number format: (XXX) XXX-XXXX

HINT: Create a dictionary that will have the area codes as keys and their respective counts as values. Add/increment the count for area codes as you iterate through the CSV file.

````python
## YOUR CODE GOES HERE ##
````

````python
## INSTRUCTION SOLUTION(S) ##
ac_counts = {}

with open('data/random_personnel.csv') as csvfile:
    
    reader = csv.reader(csvfile)
    
    data = []
    
    for row in reader:
        data.append(row)
        
    for row in data[1:]:  ## the first row is the header
        phone_num = row[1]
        area_code = phone_num[1:4]
        if area_code in ac_counts:
            ac_counts[area_code] += 1
        else: 
            ac_counts[area_code] = 1

ac_counts
````
